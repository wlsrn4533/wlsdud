```python
d_encoding = pd.get_dummies(data1['구별'])
data1 = data1.drop('구별',axis=1)
data1 = data1.join(d_encoding)
d_encoding.head(5)
#구별 원-핫 인코딩
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
X= data1[data1.columns.difference(['구별','검거 합계'])]
y = data1['검거 합계']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=19)
```

```python
lr = linear_model.LinearRegression()
model = lr.fit(X_train,y_train)
print(lr.coef_)
#회귀 분석후 학습된 계수 출력
```

```python
X_train = sm.add_constant(X_train)
model = sm.OLS(y_train,X_train).fit()
model.summary()
#라이브러리로 회귀분석 수행
```

```python
plt.rcParams['figure.figsize'] = [20,16]
coefs = model.params.tolist()
coefs_series = pd.Series(coefs)

x_labels = model.params.index.to_list()

ax = coefs_series.plot(kind='bar')
ax.set_title('feature_coef_graph')
ax.set_xlabel('x_features')
ax.set_ylabel('coef')
ax.set_xticklabels(x_labels)
#변수명을 리스트로 변환후 출력함
```

```python
X = data1[data1.columns.difference(['발생 합계','검거 합계'])]
y = data1['검거 합계']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size= 0.2,random_state=19)

lr = linear_model.LinearRegression()
model = lr.fit(X_train,y_train)

print(model.score(X_train,y_train))
print(model.score(X_test,y_test))
#학습 데이터와 테스트 데이터를 분리후 회귀분석 모델을 학습하고 평가함 > 과적합
```

```python
y_predictions = lr.predict(X_train)
print(sqrt(mean_squared_error(y_train,y_predictions)))
y_predictions = lr.predict(X_test)
print(sqrt(mean_squared_error(y_test,y_predictions)))
# 회귀분석 모델 평가
```
